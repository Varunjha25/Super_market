#### Exploratory Data analysis - Retail 

### In this Data set  we can use Python and various libraries to perform the analysis.

#### Import necessary libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

####  Load the dataset into a Pandas DataFrame

```python
data = pd.read_csv('SampleSuperstore.csv')
```

#### Explore the structure of dataset

```python
data.head()
```

```python
data.info()
```

```python
data.describe() #discriptive Statistic
```

```python
# Find the null  values in dataset 
```

```python
data.isnull().sum() # you can see there is zero null values
```

#### Calculate profit margin

```python
data['Profit margin'] = data['Profit']/data['Sales']
data.head(60)
```

#### Overall Sales and profit

```python
Overall_profit_margin = data['Profit margin'].mean()
```

```python
print(f"Overall Profit Margin: {Overall_profit_margin:.2%}")
```

#### Category and Sub-Category Analysis

```python
Cat_profit = data.groupby('Category')['Profit margin'].mean()
```

```python
Category_profit = Cat_profit.sort_values()
```

```python
print(Category_profit)
```

### Discount analysis

###### Now we use seaborn and matplotlib library to create scatter plot to analyze relation between discount and profit margin

```python
sns.scatterplot(x = 'Discount', y = 'Profit margin', data = data)
plt.title('Discount vs Profit margin')
plt.show()
```

### Mode-of-shipment

```python
shipment_profit = data.groupby('Ship Mode')['Profit margin'].mean().sort_values()
```

```python
print("Mode of Shipment-wise Profit Margins:")
print(shipment_profit)
```

### Segment-wise Analysis


```python
Segment_profit = data.groupby('Segment')['Profit margin'].mean().sort_values()
```

```python
print('Segment-wise Profit margin')
print(Segment_profit)
```

### Region-wise Analysis

```python
Region_profit = data.groupby('Region')['Profit margin'].mean().sort_values()
```

```python
print('Region-wise Profit margin')
print(Region_profit)
```

### Quantity vs Profit

##### Again we use Both library to create scatter plot to analyze between Quantity vs Profit

```python
sns.scatterplot(x = 'Quantity', y = 'Profit margin', data = data)
plt.title('Quantity vs Profit')
plt.show()
```

### Bar plot for Category vs. Profit margin

```python
plt.figure(figsize=(5, 5))
sns.barplot(x='Category', y='Profit margin', data=data)
plt.title('Average Profit for Each Category')
plt.xlabel('Category')
plt.ylabel('Profit margin')
plt.show()
```

### Bar plot for Ship Mode vs. Profit margin

```python
plt.figure(figsize=(5, 5))
sns.barplot(x='Ship Mode', y='Profit margin', data=data)
plt.title('Average Profit for Ship mode')
plt.xlabel('Ship mode')
plt.ylabel('Profit Magrin')
plt.show()
```

### Bar plot for Segment vs. Profit 

```python
plt.figure(figsize=(5, 5))
sns.barplot(x='Segment', y='Profit', data=data)
plt.title('Average Profit of Each Segment ')
plt.xlabel('Segment')
plt.ylabel('Profit Margin')
plt.show()
```
